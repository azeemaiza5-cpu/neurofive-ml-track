```python
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

df.info()
df.describe()
df.head()
```

```python
print("Shape:", df.shape)
print("\nMissing values:\n", df.isnull().sum())
print("\nColumn types:\n", df.dtypes)
```

```python
This dataset has 891 rows and 12 column. Age(177) and Cabin(687) have the most significant missing values and Embarked has only 2 missing values. Numerical columns include Age, Fare, SibSp and Parch whereas, categorical columns include Sex, Embarked and Pclass. The average passenger age is around 30 yrs and the survival rate is about 38%. This suggests class and fare likely played a role in survival. 
```
